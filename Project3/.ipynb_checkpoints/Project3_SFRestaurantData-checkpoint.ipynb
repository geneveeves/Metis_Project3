### Read in Business License Data

```python
business_locs = pd.read_csv('Registered_Business_Locations_-_San_Francisco.csv')
```

```python
business_locs.info()
```

```python
set(business_locs['NAICS Code Description'])
```

**Want to drop all business listings that are not restaurants**
 - aka keep only 'Food Services' 

```python
business_locs = business_locs.loc[business_locs['NAICS Code Description'] == 'Food Services']
```

```python
business_locs.info()
```

```python
business_locs = business_locs[['Location Id', 'Business Account Number', 
                               'DBA Name', 'Street Address', 'Source Zipcode', 
                               'Business Start Date', 'Business End Date', 
                               'Location Start Date', 'Location End Date']]
```

**Only keep observations with Zipcodes included**

```python
business_locs = business_locs[pd.notnull(business_locs['Source Zipcode'])]
```

```python
#Convert Zipcodes to int
business_locs['Source Zipcode'] = business_locs['Source Zipcode'].apply(lambda x: int(x))
```

```python
business_locs.info()
```

```python
business_locs.head()
```

**To-do:**
 
 - add column for duration LOCATION has been open
 - add column for SUCCESSFUL LOCATION

Note: business end date vs location end date matters --> location end date means they might have just moved.  Pretty sure business end date corresponds to location closing due to going out of business, not just moving, so we'll focus on those  
 

*Exclude newly-opened businesses (those open <2 yrs)*

```python
#Convert non-null Location & Business Start & End Dates to datetime objects
business_locs['Business Start Date'] = pd.to_datetime(business_locs['Business Start Date'], format='%m/%d/%Y')
business_locs['Business End Date'] = pd.to_datetime(business_locs['Business End Date'], format='%m/%d/%Y')

business_locs['Location Start Date'] = pd.to_datetime(business_locs['Location Start Date'], format='%m/%d/%Y')
business_locs['Location End Date'] = pd.to_datetime(business_locs['Location End Date'], format='%m/%d/%Y')

```

```python
#Add column for 'Currently Open' to distinguish closed locations from open ones
business_locs.loc[business_locs['Location End Date'].isnull(), 'Currently Open'] = 1
business_locs.loc[business_locs['Location End Date'].notnull(), 'Currently Open'] = 0
```

```python
#Determine length of time location has been open
```

```python
#Length of time open for closed businesses
business_locs.loc[business_locs['Location End Date'].notnull(), 'Years Open'] = ((
    business_locs['Location End Date'] - business_locs['Location Start Date']).dt.days)/365
```

```python
#Length of time open for current businesses
now = dt.datetime.now()

business_locs.loc[business_locs['Location End Date'].isnull(), 'Years Open'] = ((
    now - business_locs['Location Start Date']).dt.days)/365
```

```python
business_locs.head()
```

```python
#Pull New Restaurants out from table (currently open, and open < 2 years)
new_restaurants = business_locs.loc[(business_locs['Currently Open'] == 1) & (business_locs['Years Open'] < 2)]
```

```python
#Closed Restaurants. Add column for 'Success' --> closed after being open longer than 2 years
#Remove closed restaurants that never opened (Years open <= 0) 
closed_restaurants = business_locs.loc[(business_locs['Currently Open'] == 0) & (business_locs['Years Open'] > 0)]
```
