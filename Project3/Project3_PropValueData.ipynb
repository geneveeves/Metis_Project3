#### Scrape table with SF zipcode/neighborhood name pairings

```python
zipcode_url = 'http://www.healthysf.org/bdi/outcomes/zipmap.htm'
```

```python
#Scrape page data
zip_nbh_tables = pd.read_html(zipcode_url, flavor='bs4')
```

```python
#Select correct table
zip_nbh_table = zip_nbh_tables[2]
```

```python
#Cleaning
zip_nbh_table.rename(columns = zip_nbh_table.iloc[1], inplace=True)
```

```python
zip_nbh_table = zip_nbh_table.iloc[2:23].drop(columns = 'Population (Census 2000)')
```

```python
zip_nbh_table.reset_index(drop=True, inplace=True)
```

```python
#Convert zipcode values from strings to integers
zip_nbh_table['Zip Code'] = zip_nbh_table['Zip Code'].apply(int)
```

```python
#Pickle cleaned df for later use
zip_nbh_table.to_pickle('./zipcode_neighborhood_table.pkl')
```

#### Read in Zillow data with Timeseries zipcode data

```python
zillow_table = pd.read_csv('Zip_time_series.csv')
```

```python
zillow_table.info()
```

```python
#Read in Zipcode/SF Neighborhood table
zip_nbh_table = pd.read_pickle('./zipcode_neighborhood_table.pkl')
```

```python
zip_nbh_table.head()
```

```python
#Inner merge Zillow data with Zipcode/SF Neighborhood table so we only keep 
#information from zipcodes in SF

zipcode_data = pd.merge(zillow_table, zip_nbh_table, how='inner', left_on=['RegionName'], right_on=['Zip Code'])
```

```python
#Now we have max 5481 observations
zipcode_data.info()
```

```python
#Select only the columns we're interested in
zipcode_data = zipcode_data[['Date', 'Zip Code', 'Neighborhood', 'ZHVI_AllHomes']]
zipcode_data.info()
```

```python
#Drop observations with null values
zipcode_data = zipcode_data.dropna()
```

```python
#Convert 'Date' column to datetime object
zipcode_data['Date'] = zipcode_data.Date.apply(lambda x: dt.datetime.strptime(x, '%Y-%m-%d'))
```

```python
#Convert 'ZHVI_AllHomes' to int
zipcode_data['ZHVI_AllHomes'] = zipcode_data.ZHVI_AllHomes.apply(lambda x: int(x))
```

```python
zipcode_data.info()
```

```python
zipcode_data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline 
%config InlineBackend.figure_formats = ['retina']
sns.set_style("white")
```

```python
plt.figure(figsize=(16, 6))
sns.lineplot(x="Date", y="ZHVI_AllHomes", hue="Zip Code", data=zipcode_data);
```
